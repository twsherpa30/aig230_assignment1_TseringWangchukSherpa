## Dictionaries

```python
# Dictionaries store key–value pairs
student = {
    "name": "Jessica",
    "course": "AIG 230",
    "year": 2026
}

print(student)
print(student["course"])
```

### Modifying Dictionaries

```python
student["grade"] = "A"
student["year"] = 2027

print(student)
```

### Looping Through a Dictionary

```python
for key, value in student.items():
    print(key, ":", value)
```

# Matrices and Nested Lists

```python
# In Python, matrices are often represented as lists of lists.
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matrix)
print(matrix[0])
print(matrix[1][2])  # row 2, column 3
```

### Looping Through a Matrix

```python
for row in matrix:
    for value in row:
        print(value, end=" ")
    print()
```

# Basic Operations

```python
a = 10
b = 3

print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a ** b)  # exponentiation
```

### Operations on Lists

```python
nums = [1, 2, 3]
nums_times_two = [n * 2 for n in nums]

print(nums_times_two)
```

# Conditional Statements (if / elif / else)

```python
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
else:
    grade = "C"

print("Grade:", grade)
```

# For Loops

```python
for i in range(5):
    print("Iteration:", i)
```

### Loop with Condition

```python
for num in range(10):
    if num % 2 == 0:
        print(num, "is even")
```

```python
# Function with Multiple Arguments

def add(a, b):
    return a + b

result = add(3, 5)
print(result)
```

# Classes and Objects

```python
# Classes allow you to model real-world objects.
class Student:
    def __init__(self, name, course):
        self.name = name
        self.course = course

    def introduce(self):
        return f"My name is {self.name} and I am enrolled in {self.course}"
```

```python
student1 = Student("Jessica", "AIG 230")
print(student1.introduce())
```
